#### 10. Pclass별 나이(Age)의 평균을 구한 뒤
##### 1. 객실등급(Pclass)이 1등급인 승객의 평균 나이를 구해서, 해당 승객 중 나이(Age)컬럼값이 비어있는 승객을 찾아 빈 나이  값을 채워준다.
##### 2. 객실등급(Pclass)이 2등급인 승객의 평균 나이를 구해서, 해당 승객 중 나이(Age)컬럼값이 비어있는 승객을 찾아 빈 나이  값을 채워준다.
##### 3. 객실등급(Pclass)이 3등급인 승객의 평균 나이를 구해서, 해당 승객 중 나이(Age)컬럼값이 비어있는 승객을 찾아 빈 나이  값을 채워준다.

```python
train.groupby('Pclass')['Age'].mean()
```

```python
train['Age(fill)'] = train['Age']
```

```python
train[['Pclass','Age','Age(fill)']].head(10)
```

```python
# 객식 등급별로 Age 컬럼의 빈칸에 등급의 평균값을 채워준다.
train.loc[(train['Age'].isnull()) & (train['Pclass'] == 1), 'Age(fill)'] = train[train['Pclass']==1]['Age'].mean()
train.loc[(train['Age'].isnull()) & (train['Pclass'] == 2), 'Age(fill)'] = train[train['Pclass']==2]['Age'].mean()
train.loc[(train['Age'].isnull()) & (train['Pclass'] == 3), 'Age(fill)'] = train[train['Pclass']==3]['Age'].mean()

train[['Pclass','Age','Age(fill)']].head(10)
```

```python
### 또다른 방법
# Pclass 별로 group화 해서 null 값에 mean 을 채워 넣는다.
train['Age(fill)'] = train['Age']
train['Age(fill)'].fillna(train.groupby('Pclass')['Age(fill)'].transform('mean'), inplace=True)
train.loc[train["Age"].isnull(), ["Pclass","Age", "Age(fill)"]].head(10)
```

```python
# 가족 수에 나 자신을 포함하기 때문에 +1을 해준다.
train['FamilySize'] = train['SibSp'] + train['Parch'] + 1

train[['SibSp','Parch','FamilySize']].head(10)
```

#### 13. 가족 수(Family Size) 컬럼을 기초로 구역(범위)를 나눠준다
##### ● 싱글(Single) - 동승한 가족 X
##### ● 핵가족(Nuclear) - 동승한가족이 나 자신을 포함 2명이상 5명 미만
##### ● 대가족(Big) - 동승한 가족이 나 자신을 포함 5명 이상

```python
train.loc[train['FamilySize'] == 1, 'FamilyType'] = 'Single'
train.loc[(train['FamilySize'] >= 2) & (train['FamilySize'] < 5), 'FamilyType'] = 'Nuclear'
train.loc[train['FamilySize'] >= 5, 'FamilyType'] = 'Big'



train[['FamilySize','FamilyType']]
```

```python
# 가족 형태 (FamilyType) 에 따른 생존률
pd.pivot_table(data=train, index='FamilyType', values='Survived')
```

```python
train.groupby('FamilyType')['Survived'].mean()
```

#### 14. Single, Nuclear, Big 을 각각 One Hot Encoding 
##### 1.FamilyType(Single) -FamilySize가 1이면 True, 아니면 False
##### 2.FamilyType(Nuclear) -FamilySize가 2이상 5미만이면 True, 아니면 False
##### 3.FamilyType(Big) -FamilySize가 5이상이면 True, 아니면 False


```python
train['FamilyType(Single)'] = train['FamilySize'] == 1
train['FamilyType(Nuclear)'] = (train['FamilySize'] >=2) & (train['FamilySize'] < 5)
train['FamilyType(Big)'] = train['FamilySize'] >= 5

train[['FamilySize','FamilyType(Single)','FamilyType(Nuclear)','FamilyType(Big)']].head(10)
```

#### 15. 이름(Name) 컬럼에서 호칭(Title) 컬럼을 추출
##### 1. , 의 앞에 있는 부분 성
##### 2. , 와  . 사이에 있는 부분은 승객의 호칭
##### 3. 마지막으로 , 뒤에 있는 부분은 이름


```python
def get_title(name):
    return name.split(',')[1].split('.')[0]

train['Title'] = train['Name'].apply(get_title)

train[['Name','Title']].head(10)
```

```python
#  호칭에 따른 생존률
pd.pivot_table(data=train, index='Title', values='Survived')
```

#### 16. 호칭(Title) 컬럼을 정리

```python
# crosstab 으로 각 호칭별 생존자의 총 인원수와 사망자의 총 인원수 출력
pd.crosstab(train['Title'], train['Survived'], margins=True)
```

```python
# Mr, Mrs, Miss, Master 가 아닌 호칭은 전부 Other라고 변환

def get_title_status(title):
    title = title.strip()
    if title == 'Mr' or title == 'Miss' or title =='Mrs' or title == 'Master':
        return title
    else:
        return 'Others'

train['Title(Status)'] = train['Title'].apply(get_title_status)

train[['Name','Title', 'Title(Status)']].tail()
```

```python
# 호칭별 생존률
pd.pivot_table(data=train, index='Title(Status)', values='Survived')
```

```python
train.groupby('Title(Status)')['Survived'].mean()
```

#### 17. 호칭(Title) 컬럼으로 결혼 유무를 파악
##### ● Mr - 결혼한 남성
##### ● Master - 결혼하지 않은 남성
##### ● Mrs 결혼한 여성
##### ● Miss 결혼하지 않은 여성

```python
train['Title'] = train['Title'].str.strip()
train.loc[train['Title'] == 'Mr', 'Married'] = True
train.loc[train['Title'] == 'Master', 'Married'] = False
train.loc[train['Title'] == 'Mrs', 'Married'] = True
train.loc[train['Title'] == 'Miss', 'Married'] = False
```
